```python
import pandas as pd
```

```python
df = pd.read_csv("Salary_Data.csv")
```

```python
independant = df[["YearsExperience"]]
dependant = df[["Salary"]]
```

```python
independant
```

```python
dependant
```

```python
!pip install scikit-learn
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(independant,dependant,test_size=0.30,random_state=0)
```

```python
x_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
y_test
```

```python
Regressor = LinearRegression()
Regressor
```

```python
Regressor.fit(x_train,y_train)
```

```python
weight = Regressor.coef_
bias=Regressor.intercept_
y_predict = Regressor.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score=r2_score(y_test,y_predict)
r2_score
```

```python
import pickle
filename = "finalized_model_SLR.sav"
pickle.dump(Regressor,open(filename,'wb'))
Loaded_model=pickle.load(open("finalized_model_SLR.sav",'rb'))
years=int(input("enter:"))
```
